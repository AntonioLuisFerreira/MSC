###### a) Taxa de Compressão

```python
def taxa_compressao(dim_ori, dim_comp):
    taxa = dim_ori/dim_comp
    print("Taxa de compressao: ", round(taxa, 3),":","1")
```

```python
original = os.path.getsize("lena.tif")
#qualidade - 80  
qualidade80 = os.path.getsize("file1.jpeg")
taxa_compressao(original, qualidade80)
#qualidade - 20 
qualidade20 = os.path.getsize("file2.jpeg")
taxa_compressao(original, qualidade20)
```

###### b) Signal Noise Ratio

```python
def Measure_SNR(x, y):
    pSinal = np.sum((x*1.0)**2) / len(x)
    erro   = x-y 
    pErro  = np.sum((erro)**2) / len(erro) 
    SNR    = 10 * np.log10(pSinal/pErro)
    return SNR
```

```python
orig  = cv2.imread("lena.tif")
file1 = cv2.imread("file1.jpeg")
file2 = cv2.imread("file2.jpeg")
#SNR - 80 
aSNR = Measure_SNR(orig, file1)
print("SNR -> (file1 de 80% qualidade):", aSNR)
#SNR - 80 
bSNR = Measure_SNR(orig, file2)
print("SNR -> (file2 de 20% qualidade):", bSNR)
```

###### c) Peak Signal Noise Ratio

```python
def PSNR(x, y):
    mse = np.mean((x - y) ** 2)
    # Se o MSE for 0, não há ruído na imagem.
    # Logo, o PSNR é irrelevante.
    if(mse == 0): return 100
    psnr = 10 * np.log10(255 / np.sqrt(mse))
    return psnr
```

```python
#PSNR - 80 
aPSNR = PSNR(orig, file1)
print("PSNR -> (file1 de 80% qualidade):", aPSNR)
#PSNR - 20 
bPSNR = PSNR(orig, file2)
print("PSNR -> (file2 de 20% qualidade):", bPSNR)
```

### Pergunta 5.

- Apresente oito imagens, cada uma com o valor de cada bit para todos os pixeis.

###### Most Significant bit:
- 128 = 10000000
- 64  = 1000000
- 32  = 100000
- 16  = 10000
- 8   = 1000
- 4   = 100
- 2   = 10
- 1   = 1

```python
def shift_bits(x, bit):
    mask = 2 ** bit & x
    mask = mask >> bit
    y = x * mask
    return y

bwArray = np.array([shift_bits(x_img_g,i) for i in range(8)])

i = 0
while(i < len(bwArray)):
    imgplot = plt.imshow(cv2.cvtColor(bwArray[i], cv2.COLOR_BGR2RGB))
    title = "< BW" + " " + str(i) + " >"
    plt.axis('off')
    plt.title(title)
    plt.show()
    i+=1
```

### Pergunta 6.

- Construa uma função que realize o algoritmo de dithering Floyd Steinberg.
- Esta função recebe uma matrix (com os pixeis em tons de cinzento) e devolve uma matrix com valores a preto e branco.

```python
def get_new_val(old_val, nc):
    return np.round(old_val * (nc - 1)) / (nc - 1)
```

```python
def dither(img, nc = 2):
    arr = np.array(img, dtype=float) / 255
    new_height = len(img)
    new_width = len(img[0])
    for ir in range(new_width):
        for ic in range(new_height):
            old_val = arr[ir, ic].copy()
            new_val = get_new_val(old_val, nc)
            arr[ir, ic] = new_val
            err = old_val - new_val
            if ic < new_width - 1:
                arr[ir, ic+1] += err * 7/16
            if ir < new_height - 1:
                if ic > 0:
                    arr[ir+1, ic-1] += err * 3/16
                arr[ir+1, ic] += err * 5/16
                if ic < new_width - 1:
                    arr[ir+1, ic+1] += err * 1/16

    carr = np.array(arr/np.max(arr, axis=(0,1)) * 255, dtype=np.uint8)
    return carr
```

```python
file3 = cv2.imread("file3.bmp")
resultado = dither(file3)
plt.figure(figsize  = (7,7))
plt.axis('off')
plt.imshow(resultado)
```

### Pergunta 8.

- Crie uma função que apresente uma imagem com dimensão de 100 × 100 pixeis como se apresenta na figura.
- O ângulo de cada sector é dado por parâmetro passado à função (o ângulo é um valor inteiro entre 0 e 360 graus).

```python
def patternCreater(ang):
    x_tam = 100
    y_tam = 100
    x = np.full((x_tam, y_tam), 255)
    
    for a in range(len(x)):
        for b in range(len(x)):
            x_aux = 0
            y_aux = 0   
            if(a >= x_tam):
                x_aux = a + x_tam / 2     
            else:
                x_aux = a - x_tam / 2
            if(b <= y_tam):
                y_aux = b - y_tam / 2
            else:
                y_aux = b + y_tam / 2
            angulo = round(np.angle(x_aux + y_aux*1j, deg = True))
            if(angulo % ang == 0):
                x[a][b] = 0
            else:
                x[a][b] = 255
            
    return x
```

```python
ang = 2
padrao = np.array(patternCreater(ang),dtype = np.uint8)
plt.title('Padrão gerado com angulo de '+ str(ang))
plt.axis('off')
plt.imshow(cv2.cvtColor(padrao, cv2.COLOR_BGR2RGB))
```
