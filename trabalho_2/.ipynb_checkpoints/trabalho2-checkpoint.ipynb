### Pergunta 1.

- Elabore uma função (gen_huff_table) que gere uma tabela com o código binário para cada símbolo de um dado conjunto, usando o método de Huffman.
- Esta função deve ter como parâmetros de entrada um conjunto de símbolos e as suas probabilidades (ou em alternativa pode usar o número de ocorrências de cada símbolo, dado pelo seu histograma).
- Também pode em alternativa gerar não uma tabela mas outra estrutura de dados com os códigos pretendidos (ex: dicionário)

```python
def simb_freq(sinal):
    dicionario = {}
    for simb in sinal:
        if simb in dicionario.keys():
            dicionario[simb] += 1
        else:
            dicionario[simb] = 1
        
    return dicionario
```

```python
def gen_huff_table(simb_freq):
    #tabela a returnar
    tabela =  [[freq, [simb, ""]] for simb, freq in simb_freq.items()]
    heapify(tabela) # Faz a organização da tabela menor freq para maior 
    while len(tabela) > 1:
        right = heappop(tabela)  # heappop - Pop e return o item mais pequeno da arvore
        #print('right = ', right)
        left = heappop(tabela)
        #print('left = ', left)
        for pair in right[1:]:  
            pair[1] = '0' + pair[1]   # adiciona 0 aos da direita
        for pair in left[1:]:  
            pair[1] = '1' + pair[1]   # adiciona 1 aos da esquerda
        heappush(tabela, [right[0] + left[0]] + right[1:] + left[1:])
    
    huffman_lista = right[1:] + left[1:]
    
    huffman_dicionario = {a[0]:(str(a[1])) for a in huffman_lista}
    
    return huffman_dicionario
```

### Pergunta 3.
- Elabore uma função (decode_huff) que dada uma sequência de bits (mensagem codificada) e a tabela do ponto 1, retorne uma sequência de símbolos (mensagem descodificada).
- Garanta que a mensagem retornada por esta função é igual à mensagem que é dada como parâmetro de entrada da função encode_huff

```python
def decode_huff(tabela,msg):
    original = ''  # variavel que irá guardar a mensagem descodificacda
    msg_temp = msg # copia da mensagem    

    key_list =list(tabela.keys())
    val_list = list(tabela.values())
    huffman_lista = val_list + key_list
    tabela = {val_list[i]:key_list[i] for i in range(len(key_list))}
    
    buffer = '' # buffer utilizado no algoritmo de Huffman
    def huff_algorithm(i):
        nonlocal original
        nonlocal tabela
        nonlocal msg_temp
        nonlocal buffer
        buffer += msg_temp[i] # adicionar um bit ao buffer
        if(buffer in tabela.keys()): # verificar se o binario contido no buffer representa uma chave do dicionario
            original += str(tabela.get(buffer)) +',' # adicionar o simbolo correspondente ao codigo binario
            buffer = '' # dar reset ao buffer
        
    # Algoritmo de descodificação
    [huff_algorithm(i) for i in range(len(msg_temp))]
    return original
```

### Pergunta 4.
- Elabore uma função (encode_table) que dada a tabela de Huffman do ponto 1, retorna uma sequência de bits correspondente à tabela codificada.
- Acrescente esta sequência binária à obtida no ponto 2.

- Simbolos para binario

```python
def simbBinary(simbolos):
    simb_in_bin = []
    for i in simbolos:
        if type(i) == str: #dá jeito só para testes intermédios
            char = '0' + bin(ord(i))[2:]
            simb_in_bin = np.append(simb_in_bin,char)
        else:
            binary = bin(i)[2:]
            while len(binary) < 8:
                binary = '0' + binary
            simb_in_bin = np.append(simb_in_bin,binary)
    return simb_in_bin
```

```python
def codeBinary(codes):
    lenghtCode_in_bin = []
    for i in codes:
        binary = bin(i)[2:]
        while len(binary) < 5:
            binary = '0' + binary
        lenghtCode_in_bin = np.append(lenghtCode_in_bin,binary)
    return lenghtCode_in_bin
```

```python
def encode_table(tabela):
    key_list = [key for key in tabela.keys()]
    value_list = [value for value in tabela.values()]
    lenght_code = [len(code) for code in value_list]

    binaryNumSimb = np.binary_repr(len(key_list)-1, width=8) 
    binarySimb = simbBinary(key_list)
    binaryCodeLenght = codeBinary(lenght_code)

    encoded_table = [binarySimb[i] + binaryCodeLenght[i] + value_list[i] for i in range(len(key_list)) ]
    encoded_table = ''.join(encoded_table) #para ficar tudo numa string
    encoded_table = binaryNumSimb + encoded_table
    return encoded_table
```

### Pergunta 5.
- Elabore uma função (write2file) que dada uma sequência de bits (mensagem codificada) e o nome do ficheiro, escreva a sequência de bits para o ficheiro.

```python
def write2file(seq_bit0,filename):
    #Colocar o padding
    num = 7
    for i in range(num):
        if (len(seq_bit0) + 3 + i )%8 == 0:
            num = i
   
    num_bin = bin(num)[2:].zfill(3)
    
    stuff = ''
    for i in range(num):
        stuff = stuff + '0'
    
    seq_bit0 = num_bin + seq_bit0 + stuff
    #seq_bit0 final já com o padding feito
    
    blk = re.findall('[01]{8}', seq_bit0) # lista com blocos de 8 digitos binarios
    a = np.array([int(c,2) for c in blk], dtype = 'uint8') # converter cada um dos blocos para uint8
    f = open(filename, 'wb') # abrir o ficheiro no modo de escrita binária
    f.write(a) # escrever no ficheiro
    f.close()
```

### Pergunta 6.
- Elabore uma função (read_file) que dado o nome do ficheiro, leia uma sequência de bits (tabela codificada e mensagem codificada) contida no ficheiro e que descodifique a tabela.
- Retorne a tabela descodificada bem como a sequência binária correspondente à mensagem codificada.

```python
def remove_padding(file):
    stuf_num = file[0:3]
    stuf_num = int(stuf_num,2)
    file = file[3:]
    tamanho_ficheiro = len(file)
    #file sem o stuff
    file = file[0:len(file)-stuf_num]
    return file
```

```python
def seperate_table_from_msg(file):
    table_size = file[0:8]
    table_size = int(table_size,2)
    #remove os 1º 8 bits
    file = file[8:]
    key_binary = []
    value = []
    for i in range(table_size+1):
        key_binary = np.append(key_binary,file[0:8])
        file = file[8:]
        value_lenght = file[0:5]
        file = file[5:]
        num = int(value_lenght,2)
        value = np.append(value, file[0:num])
        file = file[num:]
    
    table = {int(key_binary[i], 2) : value[i] for i in range(table_size+1)}
   
    return table, file
```

```python
def read_file(filename):
    f = open(filename, 'rb') # modo de leitura byte a byte
    array = np.fromfile(filename, dtype = 'uint8') # cirar um array de numeros uint8
    array_bin = [format(d, '08b') for d in array] # criar um array com entradas de 8 digitos binarios
    block = ''.join(array_bin) # juntar todos os elementos do array numa unica string
    block = remove_padding(block)
    tabela = seperate_table_from_msg(block)[0]
    seq = seperate_table_from_msg(block)[1]
    return tabela, seq
```

### Pergunta 7.
- Teste as funções elaboradas usando para o efeito os seguintes ficheiros com diferentes tipos de média. Adicionalmente, pode usar mais outros ficheiros à sua escolha que achar pertinentes

- O método get_probs recebe o sinal.
- Com o auxílio do método simb_freq retorna um novo dicionário com as probabilidades de cada símbolo.

```python
def get_probs(msg):
    total = len(msg)
    dic = simb_freq(msg)
    key_list = list(dic.keys())
    val_list = list(dic.values())
    val_list = [i/total for i in val_list]
    lista = val_list + key_list
    new_dic = {key_list[i]:val_list[i] for i in range(len(key_list))}
    return new_dic # retornar o dicionario
```

- O método num_medio_bits faz a média ponderada entre a tabela descodificada e o dicionário que contém as probabilidades

```python
def num_medio_bits(tabela_cod, tabela_prob): # media ponderada
    result = 0
    for elem in tabela_cod:
        result += (len(tabela_cod[elem])*tabela_prob[elem])
    return result
```

- O método calcula a entropia usando o dicionário que contém as probabilidades

```python
def entropia(probs): # entropia da fonte
    result = 0
    for elem in probs:
        result += probs[elem]*np.log2(probs[elem])
    return -result
```

- O método eficiencia faz a devisão entre a entropia e o número médio de bits

```python
def eficiencia(entropia, n_bits): # eficiencia do codigo
    return entropia/n_bits
```

## Teste à Imagem
- Combina todos os métodos criados até então

```python
def testa_imagem(pathname,ext, gray):
    # ler a imagem
    original = [] 
    if gray:
        original = cv2.imread(pathname+ext,cv2.IMREAD_GRAYSCALE)
    else:
        original = cv2.imread(pathname+ext)
        original = cv2.cvtColor(original, cv2.COLOR_BGR2RGB)
        
    imagem = original.ravel() 
    freqImagem = simb_freq(imagem) 
    
    t0 = time()
    tabela_codigo = gen_huff_table(freqImagem) # gerar o código de Huffman
    t1 = time()
    print("Tempo de formação da tabela:", t1-t0)
    
    msg_encod = encode_huff(imagem, tabela_codigo) # codificar o ficheiro
    table_encod = encode_table(tabela_codigo)
    msg_encod = table_encod + msg_encod
    write2file(msg_encod, pathname) # garvar o ficheiro
    t2 = time()
    print("Tempo de codificação:", t2-t1)
    
    tabela, msg_encod1 = read_file(pathname)  # ler o ficheiro codificado e tratar informação
    msg_decod = decode_huff(tabela,msg_encod1)# descodificar o ficheiro     
    t3 = time()
    print("tempo de Descodificação:", t3-t2)
    
    msg_decod = np.fromstring(msg_decod, dtype=int, sep=',')
    
    size_ini = path.getsize(pathname+ext) # dimensao do ficheiro inicial
    size_end = path.getsize(pathname)     # dimensao do ficheiro resultante da codificação
    print("Taxa de Compressão: ", 1.* size_ini / size_end) # calcular a taxa de compressao
    print("Msg Descodificada = Msg Inicial? ",np.array_equal(msg_decod,imagem)) # verificar se as duas mensagens são iguais

    probsImagem = get_probs(imagem)
    
    # calcular numero medio de bits por simbolo
    n_medio_bits = num_medio_bits(tabela, probsImagem) 
    # calcular a entropia da fonte
    ent = entropia(probsImagem)            
    # calcular a eficiencia do codigo Huffman
    ef = eficiencia(ent, n_medio_bits)           
    print("Entropia da fonte:", ent)
    print("Numero Medio de bits por símbolo:", n_medio_bits)
    print("Eficiência :", ef)
    
    #Plot da Imagem 
    img = msg_decod
    if gray:
        img = img.reshape(512,512)
        plt.imshow(img, cmap='gray')
        plt.title('Imagem descodificada.')
        plt.show()
        plt.imshow(original, cmap='gray')
        plt.title('Imagem original.')
        plt.show()
    else:

        img = img.reshape(512,512,3)
        plt.imshow(img)
        plt.title('Imagem descodificada.')
        plt.show()
        plt.imshow(original)
        plt.title('Imagem original.')
        plt.show()
```
