### <center> Binary Converter </center>
- Número para binário

```python
def toBin(num): 
    if num > 0 :
        return str(bin(num))[2:]
    return str(bin(num))[3:]
```

### <center> Complemento ou Binário </center>
- Retorna o complemento caso número seja negativo.

```python
def get_comp_or_bin(num):
    if num < 0:
        return ''.join('1' if x == '0' else '0' for x in toBin(num))
    return toBin(num)
```

### <center> Quality Factor </center>

```python
def quality_factor(q):
    if(q<=50):
        factor = 50.0/q
    else:
        factor = 2.0 - (q*2.0)/100.0
    return factor 
```

### <center> SNR </center>
- Ps = Potência do sinal
- Pe = Potênica do erro

$SNR = 10 \log_{10} \frac{Ps}{Pe} $

```python
def SNR(x, y): 
    pSinal = np.sum((x*1.0)**2) / len(x)
    erro = x - y 
    pErro = np.sum((erro) ** 2) / len(erro) 
    SNR = 10 * np.log10(pSinal / pErro)
    return SNR
```

### <center> Taxa de Compressão </center>
- Tc = Taxa de Compressão
- Io = Imagem original
- Ic = Imagem comprimida

$Tc = \frac{Io}{Ic}$

```python
def taxa_compressao(pathName, ext, binaryFile):
    size_ini = path.getsize(pathName+ext) 
    size_end = path.getsize(binaryFile) 
    return 1.0* size_ini / size_end 
```

### <center> Stuffing </center>

##### Cabeçalho consiste:
- 16bits + 16bits + 6stuffedInfo + qualidade

- O método image_stuffing consiste em extender a imagem caso ela não seja divísel por 8.
- A imagem é extendida copiando a última coluna/linha o número de vezes necessária. 
- Retornando a imagem extendida e o seu cabeçalho(informações da imagem).

```python
def image_stuffing(img):
    linhas = img.shape[0]
    colunas = img.shape[1]
    
    bits_info = 3
    bites_shape_info = 16
    
    linha_adx = 8 - (linhas % 8)
    coluna_adx = 8 - (colunas % 8)
    linha_adx = 0 if linha_adx == 8 else linha_adx
    coluna_adx = 0 if coluna_adx == 8 else coluna_adx
    Linha_adx = bin(linha_adx)[2:].zfill(bits_info)
    Coluna_adx = bin(coluna_adx)[2:].zfill(bits_info)
    
    info = Linha_adx + Coluna_adx
    qualidade = '0000000'
    # 16bits + 16bits + 6stuffedInfo + qualidade -> Cabeçalho
    header = bin(linhas)[2:].zfill(bites_shape_info) + bin(colunas)[2:].zfill(bites_shape_info) + info + qualidade
    image_stuffed = np.pad(img,((0,linha_adx),(0,coluna_adx)),'edge')
    return image_stuffed, header
```

- O método rem_image_stuffing recebe a imagem extendida e o seu cabeçalho.
- Interpretando o cabeçalho retorna a imagem sem o stuff.

```python
def rem_image_stuffing(img,header):
    num = len(header)
    stuffing_info = header[num-13:num-7]
    linhas  = int( stuffing_info[0:3], 2)
    colunas = int( stuffing_info[3:] , 2)
    img = np.delete(img, range(img.shape[0] - linhas, img.shape[0]), 0)
    img = np.delete(img, range(img.shape[1] - colunas, img.shape[1]), 1)
    return img
```

### <center> 8*8 Blocks </center>
- Divide a imagem em blocos 8x8.

```python
def divide_block(img, iteracao = 8):
    nBlocks= int(img.shape[0]*img.shape[1]/(iteracao**2)) 
    blocks = np.zeros([nBlocks, iteracao, iteracao], dtype=np.uint8)
    idx = 0
    for i in range(0, img.shape[0], iteracao):
        for j in range(0, img.shape[1], iteracao):
            block = img[i:i+iteracao, j:j+iteracao]
            blocks[idx] = block
            idx += 1
    return blocks, img.shape   
```

- O método add_blocks junta todos os blocos e retorna a imagem.

```python
def add_blocks(blocks, shapeImg):
    img = np.zeros((shapeImg[0],shapeImg[1]), dtype=np.uint8)
    idx = 0
    for i in range(0, shapeImg[0], 8):         
        for j in range(0, shapeImg[1], 8):
            img[i:i+8, j:j+8] = blocks[idx]
            idx += 1
    return img
```

```python
stuffed_divided, shapeImg = divide_block(stuffed)
added_block = add_blocks(stuffed_divided, shapeImg)
np.array_equal(added_block,stuffed)
```

### <center> DCT </center>
- Consiste em cada bloco cálcular a transformada discreta de cosseno.

```python
def DCT(blocks):
    # Conversão para float
    dct_blocks = np.float32(blocks)/255.0 
    # dct função
    dct = [cv2.dct(i) for i in dct_blocks]      
    # Passar para int
    dct = [i*255.0 for i in dct]     
    return dct
```

### <center> iDCT </center>
- Faz o processo inverso

```python
def iDCT(decoded_blocks):
    # Conversão para float
    blocks = np.float32(decoded_blocks)/255.0  
    # idct função
    idct = [cv2.idct(i) for i in blocks]
    # Passar para int
    idct = [i*255.0 for i in idct]
    idct = [np.around( np.clip(i, 0, 255) ) for i in idct]
    return idct
```

### <center> Quantificação </center>
- Quantifica cada bloco consoante a qualidade.

```python
qualidades = np.array([5,25,35,45,55,75,85])

def quantific(blocks, qualidade, header):
    info_Q = 7
    qual = toBin(qualidade).zfill(info_Q)
    tmh_header = len(header)
    new_header = header[:tmh_header-7] + qual
    q = quality_factor(qualidade)
    #quantificação do sinal
    resultado = np.array([np.around(block/(Q*q)) for block in blocks])
    return resultado , new_header
```

```python
for q in qualidades:
    blocos_Q, new_header = quantific(blocos_dct, q, header)
    rec_Q = add_blocks(blocos_Q, shapeImg)
    rem_image_stuffing(rec_Q, new_header)
```

- getDCs consiste em retornar o primeiro pixel do bloco.

```python
def getDCs(blocosQ):
    result = np.array([[bloco[0][0]] for bloco in blocosQ])
    return result
```

- getAcs na extração de todos os ACs colocando-os num array

```python
def getACForEachBlock(bloco):
    counter_0 = 0
    ac_array  = []
    #começa a 1 saltando o DC
    for i in range(1, len(bloco)-1):
        valor = bloco[i]*1.0
        if(valor == 0):
            if(counter_0 == 15):
                # acrescentar um (15,0) sempre que contamos 15 zeros seguidos
                ac_array.append((counter_0, 0))
                counter_0 = 0
            else:
                counter_0 += 1
        else:
            # tuplo (numero_zeros, valor)
            ac_array.append((counter_0, valor))         
            counter_0 = 0
                
    for i in range(len(ac_array) - 1, -1, -1):          
        # tirar todos os (15,0) desnecessarios
        if(ac_array[i] == (15,0)):
            ac_array.pop(i)
        else:
            break
    # fim de bloco (EOB)            
    ac_array.append((0, 0))
    return ac_array    
```

```python
def getACs(blocosQ):
    zz = np.argsort(ind_zz)
    bloco  = [blocosQ[index].reshape((64), order='F')[zz] for index in range(len(blocosQ))]
    result = [getACForEachBlock(i) for i in bloco]

    return result
```

### <center> Desquantificação </center>

```python
def rebuild(DCs, ACs):
    blocos = np.empty( (len(DCs), 8, 8 ), dtype = np.float32) # shape da imagem original com stuffing
    
    def reconstroi_block(DCs, ACs, index):
        
        bloco    = np.zeros((64), dtype = np.float32)
        posAtual = 0
        bloco[posAtual] = DCs[index][0]
        posAtual += 1
        for i in range(len(ACs[index])):
            posAtual += ACs[index][i][0]
            bloco[posAtual] = ACs[index][i][1]
            posAtual += 1
            
        bloco8x8 = np.zeros((8,8), dtype=np.float32) # criacao de um bloco
        
        def zig_zag_to_blk(i):
            nonlocal bloco
            nonlocal bloco8x8
            valor = bloco[i]
            pos = np.where(zigzag == i) # encontra a posicao do indice i na matriz de zigzag (1D)
            bloco8x8[pos[0][0], pos[1][0]] = valor
            
        [zig_zag_to_blk(i) for i in range(len(bloco))]
        
        blocos[index] = bloco8x8 # colocar o bloco de 8 no array de blocos
    
    [reconstroi_block(DCs, ACs, index) for index in range(len(DCs))]
    
    return blocos
```

```python
def desquantific(blocosRB, fator_q):
    q = quality_factor(fator_q)
    result = np.array([bloco*(Q*q) for bloco in blocosRB]) # aplicada a cada um dos blocos
    return result
```

```python
snrs = []
snrs_tabela = []
for q in qualidades:
    blocos_Q, new_header = quantific(blocos_dct, q, header)
    DCs = getDCs(blocos_Q)
    ACs = getACs(blocos_Q)
    blocos_RB = rebuild(DCs, ACs)
    blocos_DQ = desquantific(blocos_RB, q)
    snrs.append(SNR(blocos_DQ, blocos_dct))
    snrs_tabela.append([q,SNR(blocos_DQ, blocos_dct)])
    blocos_iDCT = iDCT(blocos_DQ)
    rec_Q = add_blocks(blocos_iDCT, shapeImg)
    sem_stuff = rem_image_stuffing(rec_Q, new_header)
    imgplot = plt.imshow(sem_stuff, cmap='gray')
    plt.figure()
    
plt.plot(qualidades, snrs)
plt.xlabel( "Qualidade" )
plt.ylabel( "SNR" )
plt.title( "SNR em função da Qualidade" )
```

### <center> DPCM </center>
- Cada código DPCM (coeficiente DC) é representado pelo par {Size, Amplitude}.
- O campo Size indica quantos bits codificam o campo Amplitude.

```python
def apply_DPCM_to_DC(DCs):
    
    def altera_valor(index):
        DC_anterior = 1.0*DCs[index - 1][0]
        valor_dpcm  = 1.0*DCs[index][0] - DC_anterior
        return valor_dpcm
    
    dc = np.array([DCs[0][0]])
    dpcm_dc = np.array([altera_valor(index) for index in range(1, len(DCs))])
    return np.append(dc, dpcm_dc)

def decomp_DPCM_to_DC(DCs_dpcm):
    DCs = np.zeros((len(DCs_dpcm), 1))
    valor = 1.0*DCs_dpcm[0]
    DCs[0][0] = valor
    for i in range(1, len(DCs_dpcm)):
        DCs[i][0] = 1.0*DCs_dpcm[i] + valor
        valor = DCs[i][0]
    return DCs
```

### <center> Codificação </center>
- Codifca a informação e coloca numa trama de bits.

```python
def cod(DCs_dpcm, ACs, bit_stream):
    
    def codifica_bloco(index):
        nonlocal bit_stream
        nonlocal ACs
        binary_info = ''
        DC       = np.int32(DCs_dpcm[index])
        AC_array = ACs[index]
        dim_DC   = len(get_comp_or_bin(DC))
        binary_info += K3[dim_DC]
        bit_stream  += ('' + binary_info)
        bit_stream  += ('' + get_comp_or_bin(DC))
            
        for i in range(len(AC_array)):
            valorBin   = get_comp_or_bin(int(AC_array[i][1]))
            num_zeros  = AC_array[i][0]
            n_bits     = len(valorBin)
            # obtem o binario da K5 do tuplo (numero zeros, numero bits valor)
            bin_K5     = K5[ ( num_zeros, n_bits ) ] 
            # adiciona o binario da k5
            bit_stream += bin_K5                     
            # adiciona o binario do valor
            bit_stream += valorBin                   
        
    [codifica_bloco(index) for index in range(len(DCs_dpcm))]
    
    return bit_stream
```

### <center> Descodificação </center>
- Consiste em pegar na trama de bits.
- Usando a tabela JPEG descodifica a informação.

```python
def returnDictKey(isK3, trama):
    if isK3:
        key_list = list(K3.keys())
        value_list = list(K3.values())
        posicao = value_list.index(trama)
        return posicao
    else:
        key_list = list(K5.keys())
        value_list = list(K5.values())
        posicao = value_list.index(trama)
        tuplo_K5 = key_list[posicao]
        return tuplo_K5
```

```python
def get_bin_if_comp(num_bits, num):
    if(len(num) > 0 and num != '0' and num_bits != '1'):
        neg = num[0] == "0"
        if neg:
            valor = -1*int(''.join('1' if x == '0' else '0' for x in num),2)
            return valor
        else:
            valor = int(num,2)
            return valor
    else:
        return -1
```

```python
def decod(bit_stream):
    sizeHeader = 45
    header = bit_stream[:sizeHeader]
    bit_stream = bit_stream[sizeHeader:]
    #img_size
    linhas  = int(header[:16],2)
    colunas = int(header[ 16:32],2)
    #stuffing
    linhas_extra  = int(header[32:35],2)
    colunas_extra = int(header[35:38],2)
    
    num_blocos = int(((linhas + linhas_extra) * (colunas + colunas_extra)) / 64)
    DCs = [[]] * num_blocos
    ACs = [[]] * num_blocos
    buffer = ""
    index  = 0
    ac_array = [] # AC array que vai ser adicionado aos ACs
    while(len(bit_stream) > 0):
        buffer += bit_stream[0]
        bit_stream = bit_stream[1:]
        if(DCs[index] == []):
            if(buffer in K3.values()):
                num_bits = returnDictKey(True, buffer)
                valor    = 0
                if(num_bits != 0):
                    DC_to_Add  = bit_stream[0:num_bits]
                    bit_stream = bit_stream[num_bits:]
                    valor = get_bin_if_comp(num_bits, DC_to_Add)
                    
                DCs[index] = valor*1.0
                buffer = ""
        else:
            if(buffer != K5[(0,0)]):
                if(buffer in K5.values()):
                    tuplo_K5 = returnDictKey(False, buffer)
                    num = bit_stream[ 0: tuplo_K5[1] ]
                    bit_stream = bit_stream[ tuplo_K5[1]: ]
                    valor = get_bin_if_comp(tuplo_K5[1], num)
                    ac_array.append( (tuplo_K5[0], valor*1.0) )
                    buffer = ""
            else:
                ac_array.append((0,0.0))
                ACs[index] = ac_array
                ac_array = []
                index += 1
                buffer = ""

    return DCs, ACs, header
```

```python
stuffed, header = image_stuffing(img_g)

blocos, dimImagem = divide_block(stuffed)
blocosDCT = DCT(blocos)
blocos_Q, new_header = quantific(blocosDCT, 50, header)

DCs = getDCs(blocos_Q)
ACs = getACs(blocos_Q)
DCs_dpcm = apply_DPCM_to_DC(DCs)
bit_stream = cod(DCs_dpcm, ACs, new_header)
DCs_d, ACs_d, header = decod(bit_stream)
print("Comparacao entre DCs/Acs Originais e DCs/Acs Desquantificados")
print('DCs -> ' + str(np.array_equal(DCs_dpcm[5:25], DCs_d[5:25])))
print('ACs -> ' + str(np.array_equal(ACs[5:25], ACs_d[5:25])))
```

### <center> Guardar num ficheiro </center>

```python
def write2file(bit_stream, fileName):
    dim_stuffing_cabecalho = 3
    num_bits_stuffing = (8 - (len(bit_stream) + dim_stuffing_cabecalho) % 8) % 8
    
    bit_stream = '0' * (dim_stuffing_cabecalho - len(toBin(num_bits_stuffing))) + toBin( num_bits_stuffing ) + bit_stream
    bit_stream += '0' * num_bits_stuffing 
    
    blk = re.findall('[01]{8}', bit_stream) 

    a = np.array([int(c,2) for c in blk], dtype = 'uint8') 
    f = open(fileName, 'wb')                               
    f.write(a)                                             
    f.close()
```

### <center> Ler de um ficheiro </center>

```python
def read2String(fileName):
    f = open(fileName, 'rb')
    array = np.fromfile(fileName, dtype = 'uint8') 
    array_bin = [format(d, '08b') for d in array] 
    bit_stream = ''.join(array_bin) 
    num_stuffing_remove = int(bit_stream[:3],2)
    
    if(num_stuffing_remove > 0):    
        bit_stream = bit_stream[3:-num_stuffing_remove] 
    else:
         bit_stream = bit_stream[3:]
            
    return bit_stream
```

```python
def parse_header(header):
    #interpretação do cabeçalho
    linhas = int(header[:16],2)
    colunas = int(header[16:32],2)
    linhas_extra = header[32:35]
    colunas_extra = header[35:38]
    q = int( header[38:],2)
    
    dims_extra = linhas_extra + colunas_extra
    resultado = [linhas + int(linhas_extra,2), colunas + int(colunas_extra,2)], dims_extra, q
    return resultado
```

```python
def codificacao(img, fator_q, nomeFicheiro):
    img, bit_stream = image_stuffing(img)
    blocos, dimImagem   = divide_block(img)
    blocosDCT           = DCT(blocos)
    DCs_ACs, bit_stream = quantific(blocosDCT, fator_q, bit_stream)
    DCs = getDCs(DCs_ACs)
    ACs = getACs(DCs_ACs)
    DCs_dpcm   = apply_DPCM_to_DC(DCs)
    bit_stream = cod(DCs_dpcm, ACs, bit_stream)
    write2file(bit_stream, nomeFicheiro)
```

```python
def descodificacao(file):
    bit_stream = read2String(file)
    DCs_dpcm, ACs, cabecalho = decod(bit_stream)
    dims_img, dims_extra, q  = parse_header(cabecalho)
    DCs = decomp_DPCM_to_DC(DCs_dpcm)
    blocosRB = rebuild(DCs, ACs)
    blocos_decoded = desquantific(blocosRB, q)
    blocos_iDCT = iDCT(blocos_decoded)
    
    img_r = add_blocks(blocos_iDCT, dims_img)
    original = rem_image_stuffing(img_r, dims_extra)
    plt.title("Imagem Descodificada")
    plt.imshow(original, cmap="gray")
    plt.figure()
    return original
```
